### Class exception

```python
class B(Exception):
    pass

class C(B):
    pass

class D(C):
    pass

for cls in [B,C,D]:
    try:
        raise cls()
    except D:
        print("D")
    except C:
        print("C")
    except B:
        print("B")
```

```python
x = "Ahmad"
y = "Bakshi"

try:
    raise Exception(x, y)
except Exception as ex:
    print("Exception caught!")
    print(ex)
    print(ex.args)
    print(type(ex.args))
    f, p = ex.args
    print(f' {f} & {p}')
    
```

### Finally block

```python
try:
    print("hello world")
    raise Exception
except Exception:
    print("Exception caught")
finally:
    print("finally blocke executed")
```

```python
# Finally break cauase to not raise RuntimeError

while True:
    try:
        try:
            raise ValueError
        except ValueError:
            print("Value Error caught")
            raise RuntimeError
        finally:
            break
    except RuntimeError:
        print("RunTimeError caught")
```

### Exception group | Herarchy exceptions

```python
try:
    raise ExceptionGroup("Mutiple things went wrong",
                        [RuntimeError(1), ValueError(1)])
except* ValueError:
    print("ValueError: Caught error")
except* RuntimeError:
    print("RuntimeError: Caught error")
           
```
