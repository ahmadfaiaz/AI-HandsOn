### Using list as Stack

```python
# Last in first out (LIFO)

stack = [3,4,5]

stack.append(6)
stack.append(7)

stack.pop()
stack.pop()

print(stack)
```

### List as Queue

```python
# Deque (double-ended queue) is a generalization of stacks and queues. 
# It allows you to add and remove elements from both ends of the queue.

# First in first out (FIFO) queue

from collections import deque

queue = deque([3,4,5])

queue.append(6)
queue.append(7)

queue.popleft()
queue.popleft()

print(queue)
```

### dict() constructor

```python
tup = (1, "Apple"), (2, "Orange"), (3, "Mango")

tup_list = [tup]

print(tup_list)
print(type(tup_list))

# This will convert data into a single string 
items = [f"{num}, {name}" for num, name in tup]

# Keep data as raw, separated datatypes
items2 = [(num, name) for num, name in tup]

print(items)
print(type(items))
print(items2)
print(type(items2))
```

```python
tup1 = (1, "Apple")
tup2 = (2, "Orange")
tup3 = (3, "Cherry")

tup_list = [tup1, tup2, tup3]

print(tup_list)

print("---------------------------------")

dic = dict(tup_list)
print(dic)
print(type(dic))
```

### Looping through dict

```python
items = {"Ahmad":123, "Bob":124, "Tom":125}

for k, v in items.items():
    print(k, v)

for tup in items.items():
    print(tup)
```

```python
items = {"Ahmad":123, "Bob":124, "Tom":125}

# enumerate is a built in func that loop through iterable keep track of the index and the value
enumerate = enumerate(items, start=0)

for indx, item in enumerate:
    print(indx, item)
```

```python
qs = ['name', 'age', 'fav-color']
ans = ['Ahmad', 31, 'Blue']

# zip is a build in func that combine multiple iterable
zip_val = zip(qs, ans)

print(zip_val)
print(type(zip_val))

print('-------------------------------------')

for val in zip(qs, ans):
    print(val)
```

### del vs clear()

```python
# del vs clear - del deletes the variable from memory, 
# while clear empties the contents of the variable but keeps the variable itself in memory.

x = (1,2,3)

del x

# print(x) # raises an error because x is deleted

x = (1,2,3)
y = [1,2,3]

# x.clear() # raises an error because tuples are immutable
y.clear() # clears the list y

print(x) # prints the tuple x
print(y) # prints an empty list
```
