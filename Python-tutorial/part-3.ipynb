### Game excercise

```python
from random import randint

print("welcome to number guess game, you have 5 turns to guess the number")

rand_num = randint(1, 10)
number_of_guess = 3

while number_of_guess > 0:
    guess = int(input("please enter your guessed number: "))

    number_of_guess = number_of_guess - 1

    if guess > rand_num:
        print("Guess is greater than random num")
    elif guess < rand_num:
        print("Guess is lower than random num")
    elif guess == rand_num:
        print("your guess is correct the rand number is: " + str(rand_num))
        break
    else:
        print("something went wrong please try again")
        break
```

### Word count exercise

```python
paragraph = input("Enter your paragraph: ")
split_str = [n for n in paragraph.split(' ') if n]
print(split_str)

word_count = {}

for word in split_str:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

print(word_count)
```
