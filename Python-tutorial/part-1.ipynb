## Control Flow Tools

### If statement

```python
# Python 3.10 introduced match-case statement

x = int(input("Please enter an integer: "))

if x < 0:
    x = 0
    print("Negative change to zero")
elif x == 0:
    print("Zero")
else:
    print("the value of x is: " + str(x))
```

### Iterate over Dictionary

```python
user_status = {'Ahmad':'active', 'Bob': 'in-progress', 'Tom':'completed'}
```

```python
# copy the dictionary to avoid changing the original dictionary while iterating
# Also items() method returns a list of a dictionary's key-value tuple pairs.

for name, status in user_status.copy().items():
    status = "inactive"
    print(name +" -> "+ status)
```

```python
for name in user_status:
    print(name +" -> "+ user_status[name])
```

```python
for name in user_status.keys():
    print(name)
```

```python
print(user_status)
print(user_status.keys())
print(user_status.values())
print(user_status.items())
```

```python
print(user_status.get('ahmad'))
print(user_status.get('unknown', 'not found')) # return 'not found' if key is not found in the dictionary
print(user_status.get('unknown')) # return None if key is not found in the dictionary
print(user_status['ahmad'])
# print(user_status['unknown']) # this will raise KeyError if key is not found in the dictionary
```

### Create a new collection

```python
user_status = {'Ahmad':'active', 'Bob': 'in-progress', 'Tom':'completed'}

active_users = {}
for name, status in user_status.items():
    if status == 'active':
        active_users[name] = status
        active_users["Bob"] = "active"
        print(active_users)
        
```
