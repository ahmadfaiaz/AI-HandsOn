## Lists

```python
# [:] use to create a copy of immutable list

num = [7,10,5,17,1] + [8,11]
num_copy = num[:]

# num_copy = num
# num_copy copy the reference of num if num changes you will see the changes in num_copy
# if copy with slice [:] this will create a new list not copy the reference; Any change in num does not reflect on the slice copy

num[0] = 10

print(num)
print(num_copy)

print("-------------------------------")

num.append(9)
num.append([99])

print("-------------------------------")

num_str = str(num)
print (num_str + " hello")

print("-------------------------------")
#remove
num_copy[:] = []
print(num_copy)

print("-------------------------------")
#Destructuring
name, age = "Ahmad", 31
print(name)
print(age)

print("-------------------------------")

a,b = 0,1
while a<10:
    # end keyword use to avoid new line after the output
    print(b, end=",")
    a,b = b, a+b


```

## Control Flow Tools

### If statement

```python
# Python 3.10 introduced match-case statement

x = int(input("Please enter an integer: "))

if x < 0:
    x = 0
    print("Negative change to zero")
elif x == 0:
    print("Zero")
else:
    print("the value of x is: " + str(x))
```

### For statement

```python
words = ['cat', 'window', 'defenestrate']

for w in words:
    print(w, len(w))
```
