### Iterate over Dictionary

```python
user_status = {'Ahmad':'active', 'Bob': 'in-progress', 'Tom':'completed'}
```

```python
# copy the dictionary to avoid changing the original dictionary while iterating
# Also items() method returns a list of a dictionary's key-value tuple pairs.

for name, status in user_status.copy().items():
    status = "inactive"
    print(name +" -> "+ status)
```

### Keyword Argument

```python
""" 
optional parameter with default value 
positional parameter with default value must be defined after the required positional parameters

"""

def foo(voltage, state="Ahmad", action="fly"):
    print(voltage, state, action)


foo(10) # positional argument
foo(voltage=20) # keyword argument
foo(voltage=30, state="John", action="walk") # keyword argument
foo(40, "Bob", "jog") # positional argument
foo(50, action="run") # mixed positional and keyword argument
```

```python
# every function has a default value associated with the parameter if you do not provide a 
# value for the parameter it will use the default value

# if item_list would be None it will create a new list for the item_list parameter 
# every time the function is called

def store_item(item, item_list=[]):
    item_list.append(item)
    return item_list

print(store_item(1))
print(store_item(2))    
print(store_item(3))
print(store_item(4, [])) # this will create a new list for the item_list parameter
print(store_item(5)) # this will use the default list for the item_list parameter and append 5 to it
```

```python
# *y is used to pass a tuple after the required positional parameter x, 
def test_f(x, *y):
    print(x)
    print(y)

test_f(1,2,3,4,5)
```

```python
# **z is used to pass a dictionary after the required positional parameter x and the tuple y
def test_f(x, *y, **z):
    print(x)
    print(y)
    print(z)

    print("-------------- the way to read the tuple -----------------")

    for y in y:
        print(y)

    print("-------------- the way to read the Dictionary -----------------")

    for key, value in z.items():
        print(f"{key} -> {value}")

 
test_f(1,2,3,4,5, a=1, b=2, c=3)
```

### Arbitrary Argument

```python
# Arbitrary Argument means that you can pass any number of positional or keyword arguments to the function 
# and it will be collected in a tuple or a dictionary respectively

def concat(*args, sep="/"):
    return sep.join(args)

print(concat("earth", "mars", "venus"))
```

### UnPacking Argument List

```python
def test_unpack(x, y, z):
    print(x)
    print(y)
    print(z)

test_unpack(*[1,2,3])
print("-------------------------------")

test_unpack(*"abc")
print("-------------------------------")

test_unpack(**{"x":1, "y":2, "z":3})
print("-------------------------------")

test_unpack(*range(3))
```
