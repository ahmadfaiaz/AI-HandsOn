### String formating

```python
import math

table = {"bob": 222, "tom": 333}
for k, v in table.items():
    print(f'{k:10} ==> {v:10d}')
```

```python
print(f'value: {math.pi:.3f}')

name = 'Ahmad'
print(f'hello world my name is: {name!r}')
```

```python
ahmad, mujeeb, bob = 31,41,18
print(f'Here is age of {ahmad=} & {mujeeb=} & {bob=}')

price = 127000000
print(f'{price:,d}')
```

```python
print('Hello my name is {} and my age is {}'.format('Ahmad', '31'))
print('Hello my name is {1} and my age is {0}'.format('Ahmad', '31'))
print('Hello my name is {name} and my age is {age}'.format(name='Ahmad', age='31'))
```

```python
dic = {'ahmad': 31, 'mujeeb':41, 'bob':18}
print('Ahmad:{ahmad}; Mujeeb:{mujeeb:d}; Bob:{bob:d}'.format(**dic))
```

```python
dic = {'ahmad': 31, 'mujeeb':41, 'bob':18}
print('Ahmad:{0[ahmad]}; Mujeeb:{0[mujeeb]}; Bob:{0[bob]}'.format(dic))
```

```python
file = open("./data/note.md", 'r', encoding="UTF-8")
print(file.read())
file.close

```

```python
try:
    with open("./data/note.md", "w", encoding="UTF-8") as file:
        # raise ValueError("Manual exception raised")
        file.write("Hello just write into the file")
finally:
    print("finally blocked executed!")
    file.close()
```

```python
with open("./data/note.md", "a", encoding="UTF-8") as file:
    file.write("\nHello just append into the file")
file.close()
```

```python
# read is the default mode you can skip it if intention is read

with open("./data/note.md", encoding="UTF-8") as file:
    print(file.read())
file.close()
```

```python
with open("./data/note.md", encoding="UTF-8") as file:
    print(file.read(40))
    print(file.read(40))
file.close
```

```python
with open("./data/note.md", encoding="UTF-8") as file:
    for line in file:
        print(line)
        print("=============================")
```

```python
with open("./data/note.md", encoding="UTF-8") as file:
    print(list(file))
file.close

    
```

```python
with open("./data/note.md", encoding="UTF-8") as file:
    print(file.readlines())
file.close
```

```python
# tell() return the current position of a file pointer(cursor) within file stream
# seek() force to cursor to a specific byte location

with open("./data/note.md", encoding="UTF-8") as file:

    file_start = file.tell()
    print(file.read(30))
    file.seek(file_start)
    print(f'*********{file.tell()}*************')
    print(file.read(30))
    file.seek(file_start)
    print(f'*********{file.tell()}*************')
    print(file.read(30))
    file.seek(file_start)
    print(f'*********{file.tell()}*************')
file.close
```

### json

```python
import json

x = [1, 'simple', 'list']

y = json.dumps(x)
print(y)
print(type(y))
```

```python
import json

x = [1, 'simple', 'list']

y = json.dumps(x)
print(y)
print(type(y))
```

```python
# dump into a file

import json

x = [1, 'simple', 'list']

with open("./data/note2.md", "w", encoding="UTF-8") as file:
    z = json.dump(x, file)
    print(z)
    print(type(z))
file.close()
```

```python
import json

with open("./data/note2.md", encoding="UTF-8") as file:
    data = json.load(file)
file.close

print(data)
print(type(data))
```
