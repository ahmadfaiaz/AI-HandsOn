##### Namespace vs Symbol_Table:
 A namespace is a runtime environment that maps names to objects, while a symbol table is a compiler-time data structure used to track scopes and identifier properties before bytecode is generated

```python
# Scope with no nonloacl keyword

foo = 10

def fun():
    foo = 30
    def in_fun():
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

```python
# Scope with nonloacl keyword
# Note: nonlocal can not bind wiht module (global) level

foo = 10

def fun():
    foo = 30
    def in_fun():
        nonlocal foo
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

```python
# Note: Global can bind wiht module level

foo = 10

def fun():
    global foo
    foo = 30
    def in_fun():
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

### Class

```python
class MyClass:
    foo = 10

    def __init__(self):
        self.foo = 20
    
    def fun():
        print("Fun method")

print(MyClass.foo)
x = MyClass()
print(x.foo)
```

### Inheritance

```python
class A:
    num = 10
    def funA(self):
        print("Hello this is class A")

class B (A):
    def __init__(self):
        print("Hello from constructor")
        self.funA()

    def funB(self):
        print("Hello this is class B")

x = B()
x.funB()

```

```python
class Person:

    def __init__(self, f_name, l_name):
        self.f_name = f_name
        self.l_name = l_name

    def print_name(self):
        print(f'{self.f_name} {self.l_name}')

class Programmer (Person):
    
    def __init__(self, f_name, l_name, fav_lang):
        Person.__init__(self, f_name, l_name)
        self.fav_lang = fav_lang

    def print_name(self):
        print(f"{self.l_name}, {self.f_name} loves {self.fav_lang}")

parent = Person("Ahmad", "Bakshi")
parent.print_name()

child = Programmer("Tom", "Cruise", "Java")
child.print_name()

```
