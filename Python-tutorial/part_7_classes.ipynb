##### Namespace vs Symbol_Table:
 A namespace is a runtime environment that maps names to objects, while a symbol table is a compiler-time data structure used to track scopes and identifier properties before bytecode is generated

```python
# Scope with no nonloacl keyword

foo = 10

def fun():
    foo = 30
    def in_fun():
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

```python
# Scope with nonloacl keyword
# Note: nonlocal can not bind wiht module (global) level

foo = 10

def fun():
    foo = 30
    def in_fun():
        nonlocal foo
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

```python
# Note: Global can bind wiht module level

foo = 10

def fun():
    global foo
    foo = 30
    def in_fun():
        foo = 20
        print(f'in_fun: {foo}')
    
    in_fun()
    print(f'fun: {foo}')

fun()
print(f'global: {foo}')
```

### Class

```python
class MyClass:
    foo = 10

    def __init__(self):
        self.foo = 20
    
    def fun():
        print("Fun method")

print(MyClass.foo)
x = MyClass()
print(x.foo)
```
